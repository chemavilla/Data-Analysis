```python
import pandas as pd
import seaborn as sns
import matplotlib as plt

url='https://raw.githubusercontent.com/softhints/Pandas-Exercises-Projects/main/data/food_recipes.csv'
df = pd.read_csv(url)
```

```python
df.info()
```

```python
df.sample(5)
```

```python
df.describe(include='object')
```

```python
df['prep_time']= df['prep_time'].str[0:2]
df['cook_time']= df['cook_time'].str[0:2]

df=df[df['prep_time'].notna()].astype({'prep_time':int})
df=df[df['cook_time'].notna()].astype({'cook_time':int})

```

```python
df.info()
```

```python
df1= df.groupby(by='prep_time').count()

df1= df1[df1["url"]>80].sort_values(by='prep_time')

ax=df1['url'].plot(kind='bar')
ax.set(xlabel="Time", ylabel="Num")
```

```python
df2= df.groupby(by='cook_time').count()

df2= df2[df2["url"]>80].sort_values(by='cook_time')

ax=df2['url'].plot(kind='bar')
ax.set(xlabel="Time", ylabel="Num")
```

```python
df12 = pd.DataFrame()
df12['prep_time'] = df1['url']
df12['cook_time'] = df2['url']
df12.plot()
```
